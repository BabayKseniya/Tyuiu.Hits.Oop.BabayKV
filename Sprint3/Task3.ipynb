Создать базовый класс Person в C#, который будет представлять информацию о человеке. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. Требования к базовому классу Person: • Атрибуты: Имя (Name), Возраст (Age), Пол (Gender). • Методы: o GetInfo(): метод для получения информации о человеке в виде строки. o SayHello(): метод для вывода приветствия от человека. Требования к производным классам:

Студент (Student): Должен содержать дополнительные атрибуты, такие как Университет (University) и Курс (Course). Метод SayHello() должен быть переопределен для добавления информации о курсе при приветствии.
Работник (Employee): Должен содержать дополнительные атрибуты, такие как Компания (Company) и Зарплата (Salary). Метод GetInfo() должен быть переопределен для включения информации о компании и зарплате.
Преподаватель (Teacher) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Предмет (Subject) и Стаж работы (Experience). Метод SayHello() должен быть переопределен для добавления информации о преподаваемом предмете при приветствии.

```
public class Person
{
    
    public string Name { get; set; }
    public int Age { get; set; }
    public string Gender { get; set; }
    
    
    public string Email { get; set; }
    public string Phone { get; set; }
    public string Address { get; set; }
    public DateTime BirthDate { get; set; }

    
    public Person() { }

    public Person(string name, int age, string gender)
    {
        Name = name;
        Age = age;
        Gender = gender;
    }

    public Person(string name, int age, string gender, string email, string phone, string address, DateTime birthDate)
    {
        Name = name;
        Age = age;
        Gender = gender;
        Email = email;
        Phone = phone;
        Address = address;
        BirthDate = birthDate;
    }

    
    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Пол: {Gender}";
    }

    public virtual void SayHello()
    {
        Console.WriteLine($"Привет! Меня зовут {Name}.");
    }

    
    public virtual void DisplayContactInfo()
    {
        Console.WriteLine($"Контактная информация: Email: {Email}, Телефон: {Phone}");
    }

    public int CalculateBirthYear()
    {
        return DateTime.Now.Year - Age;
    }

    public virtual void Work()
    {
        Console.WriteLine($"{Name} выполняет общие задачи.");
    }
}


public class Student : Person
{
    
    public string University { get; set; }
    public int Course { get; set; }
    public string Faculty { get; set; }
    public double AverageGrade { get; set; }
    public string StudentId { get; set; }

    public Student() { }

    public Student(string name, int age, string gender, string university, int course)
        : base(name, age, gender)
    {
        University = university;
        Course = course;
    }

    
    public override string GetInfo()
    {
        return base.GetInfo() + $", Университет: {University}, курс: {Course}";
    }

    public override void SayHello()
    {
        Console.WriteLine($"Привет! Я студент {Name}, учусь на {Course} курсе в {University}.");
    }

    
    public void Study()
    {
        Console.WriteLine($"{Name} изучает предметы на {Course} курсе.");
    }

    public void TakeExam(string subject)
    {
        Console.WriteLine($"{Name} сдает экзамен по {subject}.");
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} готовится к занятиям и выполняет домашние задания.");
    }
}


public class Employee : Person
{
    
    public string Company { get; set; }
    public decimal Salary { get; set; }
    public string Position { get; set; }
    public string Department { get; set; }
    public int YearsInCompany { get; set; }

    public Employee() { }

    public Employee(string name, int age, string gender, string company, decimal salary)
        : base(name, age, gender)
    {
        Company = company;
        Salary = salary;
    }

    
    public override string GetInfo()
    {
        return base.GetInfo() + $", Компания: {Company}, Зарплата: {Salary} рублей, Должность: {Position}";
    }

    public override void SayHello()
    {
        Console.WriteLine($"Здравствуйте! Я {Name}, работаю в компании {Company}.");
    }

    
    public void WorkOnTask(string task)
    {
        Console.WriteLine($"{Name} работает над задачей: {task}");
    }

    public void AttendMeeting()
    {
        Console.WriteLine($"{Name} участвует в совещании.");
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} выполняет рабочие обязанности в компании {Company}.");
    }

    
    public void Work(int hours)
    {
        Console.WriteLine($"{Name} работает {hours} часов сегодня.");
    }
}


public class Teacher : Person
{
    
    public string Subject { get; set; }
    public int Experience { get; set; }
    public string AcademicDegree { get; set; }
    public string[] Courses { get; set; }
    public bool IsProfessor { get; set; }

    public Teacher() { }

    public Teacher(string name, int age, string gender, string subject, int experience)
        : base(name, age, gender)
    {
        Subject = subject;
        Experience = experience;
    }

    
    public override string GetInfo()
    {
        return base.GetInfo() + $", Предмет: {Subject}, Стаж: {Experience} лет, Ученая степень: {AcademicDegree}";
    }

    public override void SayHello()
    {
        Console.WriteLine($"Добрый день! Я преподаватель {Name}, преподаю {Subject} уже {Experience} лет.");
    }

    
    public void Teach()
    {
        Console.WriteLine($"{Name} проводит занятие по {Subject}.");
    }

    public void GradeStudents()
    {
        Console.WriteLine($"{Name} проверяет работы студентов.");
    }

    public void PrepareLesson()
    {
        Console.WriteLine($"{Name} готовится к следующему уроку.");
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} преподает {Subject} и готовится к занятиям.");
    }

    
    public void Work(string specificTask)
    {
        Console.WriteLine($"{Name} работает над: {specificTask}");
    }
}

// Generic класс
public class PersonManager<T> where T : Person
{
    private List<T> _people;

    public PersonManager()
    {
        _people = new List<T>();
    }

    public void AddPerson(T person)
    {
        _people.Add(person);
        Console.WriteLine($"Добавлен: {person.Name}");
    }

    public void RemovePerson(T person)
    {
        _people.Remove(person);
        Console.WriteLine($"Удален: {person.Name}");
    }

    public void DisplayAllInfo()
    {
        Console.WriteLine("\n Информация о всех людях ");
        foreach (var person in _people)
        {
            Console.WriteLine(person.GetInfo());
        }
    }

    public void MakeAllSayHello()
    {
        Console.WriteLine("\n Приветствия всех людей ");
        foreach (var person in _people)
        {
            person.SayHello();
        }
    }

    public List<T> FindByAge(int minAge, int maxAge)
    {
        return _people.FindAll(p => p.Age >= minAge && p.Age <= maxAge);
    }

    public int Count => _people.Count;
}


        
        var student = new Student("Кристина Зарева", 20, "Женский", "ТИУ", 3)
        {
            Email = "kris@mail.ru",
            Phone = "+79991234567",
            Faculty = "Филологический",
            AverageGrade = 4.8,
            StudentId = "00001"
        };

        var employee = new Employee("Кирилл Осинцев", 35, "Мужской", "ООО ТехноПро", 85000 )
        {
            Email = "kiriil@gmail.com",
            Phone = "+74951234567",
            Position = "Разработчик",
            Department = "IT",
            YearsInCompany = 5
        };

        var teacher = new Teacher("Ольга Владимировна", 45, "Женский", "Математика", 15)
        {
            Email = "olga@university.ru",
            Phone = "+74999876543",
            AcademicDegree = "Кандидат наук",
            Courses = new string[] { "Алгебра", "Математический анализ" },
            IsProfessor = true
        };

        
        Person[] people = { student, employee, teacher };

        foreach (var person in people)
        {
            person.SayHello(); 
            Console.WriteLine(person.GetInfo()); 
            person.Work(); 
            Console.WriteLine();
        }
        var personManager = new PersonManager<Person>();
    
        personManager.AddPerson(student);
        personManager.AddPerson(employee);
        personManager.AddPerson(teacher);
        personManager.DisplayAllInfo();
        personManager.MakeAllSayHello();


        employee.Work(8); 
        teacher.Work("научная статья"); 

      
        student.Study();
        student.TakeExam("Программирование");
        
        employee.WorkOnTask("Разработка нового модуля");
        employee.AttendMeeting();
        
        teacher.Teach();
        teacher.GradeStudents();
        teacher.PrepareLesson();

    
        Console.WriteLine("\n Поиск людей в возрасте 25-40 лет");
        var peopleInAgeRange = personManager.FindByAge(25, 40);
        foreach (var person in peopleInAgeRange)
        {
            Console.WriteLine($"Найден: {person.Name}, возраст: {person.Age}");
        }

        Console.WriteLine($"\n Общее количество людей в системе: {personManager.Count}");
```
